Medidas de Dispersão

```python
# Variância, Desvio Padrão e Amplitude.
```

```python
""" A variância é uma métrica estatística que descreve o quanto cada número em
um conjunto de dados varia da média do conjunto. Em outras palavras, é uma
medida de dispersão ou espalhamento dos dados. Se a variância for pequena,
 significa que os números estão perto da média do conjunto; se for grande,
 significa que os números estão mais espalhados em relação à média."""
```

```python
import matplotlib.pyplot as plt

def plot_variance(data):
    # Cálculo da média
    mean = np.mean(data)

    # Cálculo da variância
    variance = np.var(data)

    plt.figure(figsize=(10, 6))

    # Histograma
    plt.hist(data,
             bins=20,
             color='skyblue',
             edgecolor='black',
             alpha=0.7)

    # Linha representando a média
    plt.axvline(mean,
                color='red',
                linestyle='dashed',
                linewidth=2,
                label=f'Média: {mean:.2f}')

    # Linhas do topo do histograma até a média
    hist, bins = np.histogram(data, bins=20)
    for i in range(len(hist)):
        plt.vlines(bins[i], 0, hist[i], color='gray',
                   linestyle='dotted', linewidth=0.5)
        plt.vlines(bins[i+1], 0, hist[i], color='gray',
                   linestyle='dotted', linewidth=0.5)

    plt.title(f"Distribuição dos Dados e Variância (Variância = {variance:.2f})")
    plt.legend()
    plt.show()
```

```python
# Gerando um conjunto de dados aleatórios
data = np.random.normal(0, 5, 1000)
#chama função
plot_variance(data)
```

```python
def plot_variance2(data):
    # Cálculo da média
    mean = np.mean(data)

    # Cálculo da variância
    variance = np.var(data)

    plt.figure(figsize=(10, 6))

    # Plot dos dados
    plt.scatter(data, [1] * len(data), color='blue', s=50, label='Dados')

    # Linha representando a média
    plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Média: {mean:.2f}')

    # Linhas do ponto até a média
    for point in data:
        plt.plot([point, mean], [1, 1.5], color='gray', linestyle='dotted', linewidth=1)

    plt.title(f"Gráfico da Variância (Variância = {variance:.2f})")
    plt.yticks([])
    plt.legend()
    plt.show()
```

```python
# Testando com um conjunto de dados exemplo
data = [2, 3, 5, 7, 11, 13, 17]
plot_variance2(data)
```

```python
""" O desvio padrão é uma métrica estatística que mede a quantidade de variação
ou dispersão de um conjunto de valores. Ele indica o quão "espalhados" estão os
dados em relação à média, e é especialmente útil para entender a variabilidade
em grandes conjuntos de dados.

Um desvio padrão baixo indica que os valores tendem a estar próximos da média
(ou valor esperado) do conjunto, enquanto um desvio padrão alto indica que os
valores estão espalhados em uma faixa mais ampla. """
```

```python
# Gerando dados aleatórios com média 50 e desvio padrão 10
data = np.random.normal(50, 10, 1000)

# Calculando a média e o desvio padrão
mean = np.mean(data)
std_dev = np.std(data)
print("Desvio padrão: ", std_dev)

plt.figure(figsize=(10, 6))

# Histograma
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Linha da média
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label=f'Média: {mean:.2f}')

# Linhas de desvio padrão
for i in range(1, 4):
    plt.axvline(mean - i*std_dev, color='green', linestyle='dotted', linewidth=1, label=f'Média - {i} desvio padrão')
    plt.axvline(mean + i*std_dev, color='green', linestyle='dotted', linewidth=1, label=f'Média + {i} desvio padrão')

plt.legend()
plt.title("Ilustração do Desvio Padrão")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Codifica rótulos (strings -> inteiros)
encoder = LabelEncoder()
y_treino_enc = encoder.fit_transform(y_treino)
y_teste_enc  = encoder.transform(y_teste)
```

```python
# Modelo com MLP
model = make_pipeline(
    StandardScaler(),
    MLPClassifier(
        hidden_layer_sizes=(100,),
        activation="relu",
        solver="adam",
        alpha=1e-4,
        learning_rate_init=1e-3,
        max_iter=500,
        early_stopping=True,
        n_iter_no_change=10,
        random_state=42
    )
)
```

```python
model.fit(X_treino, y_treino_enc)
```

```python
# Predição
y_pred = model.predict(X_teste)
accuracy = accuracy_score(y_teste_enc, y_pred)
print("Accuracy:", accuracy)
```

```python
# Decodifica previsões de volta para rótulos originais
y_pred_labels = encoder.inverse_transform(y_pred)
print(classification_report(y_teste, y_pred_labels))
```
